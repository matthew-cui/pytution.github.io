## Locate Records

Two major methods mention here, loc and iloc
* loc is use for location by column name and row label
* iloc is integer-location based indexing for selection by position


```
# sample of DataFrame.iloc
# Purely integer-location based indexing for selection by position
import pandas as pd
import numpy as np

d = {'col1': [1,2,4,5], 'col2': [3, 4,4,5],'col3': [5,6,4,5]}
index = ['row1','row2','row3','row4']
df = pd.DataFrame(data=d,index=index)
print(df.head())
# it will retrieve the 1st and 3rd rows and 0th and 2nd cols cells
rs = df.iloc[[1,3],[0,2]]
print("##print rs...")
print(rs)

# if the index is not found, then return IndexError
try:
  rsx = df.iloc[7]
except IndexError as err:
  print("record not found")
# col1 is the column name and 3 is the index value
# If the record not found, it will raise KeyError
try:
  print('##print col1 row3 value')
  print(rs['col1']['row3'])
except KeyError as err:
  print("record not found")
```

## Retreive Cell Value
Simpilar to the location, there are two methods at and iat to retreive the cell value from dataframe
* at is using the column name and row label
* iat is integer based retreive method


```
# retrieve the value from single cell also can use at method
# using the index label name and column name (row/column)
print('##print row2 col1 value')
print(rs.at['row2','col1'])
# using the index of column and row, i.e. return the (0,0) of result set
print(rs.iat[0,0])
```

## Iterates DataFrame
The below methods can be referred to use for dataframe iteration 
* *iteritems* iterates by columns and returns series
* *iterrows* iterates by rows and returns series

```
import pandas as pd
import numpy as np

# sample data
d = {'col1': [1,2,4,5], 'col2': [3, 4,4,5],'col3': [5,6,4,5]}
index = ['row1','row2','row3','row4']
df = pd.DataFrame(data=d,index=index)
print("display sample data frame")
print(df)

for label,item in df.iteritems():
  print(f"Label      :{label}")
  print(f"Column Series {item}")

for index,row in df.iterrows():
  print(f"Index   :{index}")
  print(f"Row Series:{row}")
```
